```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["ana", "bruno", "carlos", "daniele", "val"],
    'salario': [4000, np.nan, 5000, np.nan, 10000]
})
#calcula a mdia salarial e substitui os nulos pela mdia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''calcula a mediana salarial e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "dia": ["segunda", "tera", "quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ["ana", "bruno", "carlos", "daniele", "val"],
    "cidade": ["so paulo", np.nan, "curitiba", np.nan, "belm"]
 })
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("no informado")
df_cidades
```

# AULA 3 - VISUALIZAO DE DADOS

```python
df_limpo.head()
```

```python
df_limpo["experience_level"].value_counts().plot(kind="bar", title = "distribuio de senioridade")
```

```python
import seaborn as sns
```

```python
sns.barplot(data = df_limpo, x= "experience_level", y= "salary_in_usd")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (8,5))
sns.barplot(data = df_limpo, x= "experience_level", y= "salary_in_usd")
plt.title("salario mdio por senioridade")
plt.xlabel("senioridade")
plt.ylabel("salario medio anual (usd)")
plt.show()
```

```python
df_limpo.groupby("experience_level")["salary_in_usd"].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby("experience_level")["salary_in_usd"].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize= (8,5))
sns.barplot(data = df_limpo, x= "experience_level", y= "salary_in_usd", order=ordem)
plt.title("salario mdio por senioridade")
plt.xlabel("senioridade")
plt.ylabel("salario medio anual (usd)")
plt.show()
```

```python
plt.figure(figsize= (10,6))
sns.histplot(df_limpo["salary_in_usd"], bins = 50, kde=True)
plt.title("distribuio dos salrios anuais")
plt.xlabel("salario em usd")
plt.ylabel("frequencia")
plt.show()
```

```python
plt.figure(figsize= (8,6))
sns.boxplot(x= df_limpo["salary_in_usd"])
plt.title("boxplot salario")
plt.xlabel("salario em usd")

plt.show()
```

```python
ordem_senioridade = ["EN", "MI", "SE", "EX"]
plt.figure(figsize= (8,6))
sns.boxplot(x= "experience_level", y= "salary_in_usd", data= df_limpo, order= ordem_senioridade)
plt.title("boxplot da distribuio da senioridade")
plt.xlabel("salario em usd")
plt.show()
```

```python
ordem_senioridade = ["EN", "MI", "SE", "EX"]
plt.figure(figsize= (8,6))
sns.boxplot(x= "experience_level", y= "salary_in_usd", data= df_limpo, order= ordem_senioridade, palette="Set2", hue= "experience_level")
plt.title("boxplot da distribuio da senioridade")
plt.xlabel("salario em usd")
plt.show()
```

```python
import plotly.express as px
```

```python
df_avg_salary_experience = df_limpo.groupby('experience_level')['salary_in_usd'].mean().reset_index()

fig = px.bar(df_avg_salary_experience,
             x='experience_level',
             y='salary_in_usd',
             title='Mdia Salarial por Nvel de Experincia',
             labels={'experience_level': 'Nvel de Experincia', 'salary_in_usd': 'Mdia Salarial (USD)'})
fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':['EN', 'MI', 'SE', 'EX']})
fig.show()
```
